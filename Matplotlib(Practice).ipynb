**Matplotlib Data Visualization **

**মডিউল ১: বেসিক Plotting এবং Line Charts**


**১. matplotlib এবং pandas import করুন এবং sales_data.csv ফাইলটি লোড করুন।**

```python
df=pd.read_csv("sales_data.csv")
df.head()
```

**২. একটি simple line plot তৈরি করুন যেখানে x-axis এ 'Month' এবং y-axis এ 'Total_Sales' দেখাবে।**

```python

month_order = ['January', 'February', 'March', 'April', 'May', 'June',
               'July', 'August', 'September', 'October', 'November', 'December']



monthly_sales = df.groupby('Month')['Total_Sales'].sum().reindex(month_order).reset_index()
plt.figure(figsize=(12, 5))
plt.plot(monthly_sales['Month'], monthly_sales['Total_Sales'], marker='o')
plt.xlabel("Month")
plt.ylabel("Total_Sales")
plt.title("Monthly Total Sales")
plt.grid(True)
```

**৩. একই plot এ দুটি line যোগ করুন - একটি 'Total_Sales' এবং অন্যটি 'Total_Profit' (different colors সহ)।**

```python
monthly_data = df.groupby('Month')[['Total_Sales', 'Total_Profit']].sum().reindex(month_order).reset_index()

plt.figure(figsize=(12, 6))
plt.plot(monthly_data['Month'],monthly_data['Total_Sales'], color="blue", marker="o", label="Total Sales")

plt.plot(monthly_data['Month'],monthly_data['Total_Profit'], color="green", marker=">", label="Total Profit")

plt.xlabel('Month', fontsize=12)
plt.ylabel('Amount (in BDT)', fontsize=12)
plt.title('Comparison: Monthly Total Sales vs Profit', fontsize=14)
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()
```

**৪. উপরের plot টিতে title 'Monthly Sales and Profit Trends', xlabel 'Month', ylabel 'Amount (BDT)', এবং legend যোগ করুন।**

```python
monthly_data = df.groupby('Month')[['Total_Sales', 'Total_Profit']].sum().reindex(month_order).reset_index()

plt.figure(figsize=(12, 6))
plt.plot(monthly_data['Month'],monthly_data['Total_Sales'], color="blue", marker="o", label="Total Sales")

plt.plot(monthly_data['Month'],monthly_data['Total_Profit'], color="green", marker=">", label="Total Profit")
plt.xlabel('Month', fontsize=12)
plt.ylabel('Amount (BDT)', fontsize=12)
plt.title('Monthly Sales and Profit Trends', fontsize=14)

plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()
```

**৫. Line style এবং marker customize করুন (যেমন: dashed line, circle markers, line width = 2)।**

```python
monthly_data = df.groupby('Month')[['Total_Sales', 'Total_Profit']].sum().reindex(month_order).reset_index()

plt.figure(figsize=(12, 6))

plt.plot(monthly_data['Month'], monthly_data['Total_Sales'],
         marker='o',linestyle='--', label='Total Sales', color='blue', linewidth=2)

plt.plot(monthly_data['Month'], monthly_data['Total_Profit'],
         marker='o',linestyle='--', label='Total Profit', color='red', linewidth=2)


plt.xlabel('Month', fontsize=12)
plt.ylabel('Amount (BDT)', fontsize=12)
plt.title('Monthly Sales and Profit Trends', fontsize=14)


plt.legend()
plt.tight_layout()
plt.show()
```

**৬. প্রতিটি product এর total sales একটি vertical bar chart এ দেখান।**

```python
df = pd.read_csv("sales_data.csv", usecols=["Product", "Total_Sales"])
Product = df['Product']
Total_Sales = df['Total_Sales']
plt.figure(figsize=(8, 6))
plt.bar(Product, Total_Sales)
plt.xlabel("Products")
plt.ylabel("Total Sales")
plt.title("Total Sales by Product")
plt.show()
```

**৭. একটি horizontal bar chart তৈরি করুন যেখানে প্রতিটি region এর total revenue দেখাবে।**

```python
df = pd.read_csv("sales_data.csv", usecols=["Region", "Total_Sales"])
plt.figure(figsize=(5, 3))
region_sales = df.groupby("Region")["Total_Sales"].sum()
plt.barh(region_sales.index, region_sales.values)
plt.xlabel("Total Revenue")
plt.ylabel("Region")
plt.title("Total Revenue by Region")
plt.show()
```
